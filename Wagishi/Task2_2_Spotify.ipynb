```python
song['streams']=song['streams'].fillna(0)
song['in_shazam_charts'] = song['in_shazam_charts'].fillna(0)
song['key'] = song['key'].fillna('unknown')
song.isnull().sum()
```

Changing the data type

```python
song['streams'] = pd.to_numeric(song['streams'], errors='coerce')
song['in_deezer_playlists'] = pd.to_numeric(song['in_deezer_playlists'], errors='coerce')
song['in_shazam_charts'] = pd.to_numeric(song['in_shazam_charts'], errors='coerce')
```

model fitting


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
x=song_encoded.drop(columns=['streams','track_name','artist(s)_name'],axis=1)
y=song_encoded["streams"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Remove redundant LinearRegression initialization
lr=LinearRegression()
```

```python
# Check for and handle infinite values in x_train and y_train
import numpy as np

# Identify rows with NaN or infinite values in x_train or y_train
finite_mask = np.isfinite(x_train).all(axis=1) & np.isfinite(y_train)

# Filter out rows with non-finite values
x_train = x_train[finite_mask]
y_train = y_train[finite_mask]

lr.fit(x_train,y_train)
```

```python
# Fill missing values in x_test with the mean of the test set
x_test = x_test.fillna(x_test.mean())

y_pred = lr.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

```python
test_indices = x_test.index
test_songs = song.loc[test_indices]

predictions_df = pd.DataFrame({
    'artist(s)_name': test_songs['artist(s)_name'],
    'track_name': test_songs['track_name'],
    'predicted_streams': y_pred
})
```
