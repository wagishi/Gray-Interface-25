````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
dataset = pd.read_csv('CC GENERAL.csv')
````

````python
dataset.shape
````

````python
dataset.head()
````

````python
dataset.drop(columns = 'CUST_ID', inplace = True)
````

````python
dataset.describe()
````

````python
dataset.isnull().sum()
````

````python
dataset = dataset.fillna(dataset.mean())
````

````python
cols = 4
num_features = dataset.shape[1]
rows = int(np.ceil(num_features / cols))

# Create subplots matrix
plt.figure(figsize=(20, rows * 3))

for i, column in enumerate(dataset.columns):
    plt.subplot(rows, cols, i + 1)
    sns.boxplot(y=dataset[column], color="skyblue")
    plt.title(column, fontsize=10)
    plt.tight_layout()

plt.suptitle("Boxplots of All Features (Outlier Detection)", fontsize=16, y=1.02)
plt.show()
````

````python
Q1 = dataset.quantile(0.08)
Q3 = dataset.quantile(0.95)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
dataset_cleaned = dataset[~((dataset < lower_bound) | (dataset > upper_bound)).any(axis=1)]

print("Original dataset shape:", dataset.shape)
print("After outlier removal:", dataset_cleaned.shape)
````

here we have around 4.25% of loss

````python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(dataset_cleaned)
````

````python
from sklearn.cluster import KMeans
sse = []
for k in range(1,10):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(X)
    sse.append(km.inertia_)
plt.plot(range(1,10), sse, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.show()
````

````python
from sklearn.cluster import KMeans
k = 6
km = KMeans(n_clusters=k, random_state=42, n_init=10)
labels = km.fit_predict(X)
# Create a copy to avoid SettingWithCopyWarning
dataset_cleaned_copy = dataset_cleaned.copy()
dataset_cleaned_copy['cluster_kmeans'] = labels
````

````python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
coords = pca.fit_transform(X)
plt.figure(figsize=(8,6))
sns.scatterplot(x=coords[:,0], y=coords[:,1], hue=labels, palette='tab10')
plt.title("K-means clusters (k={})".format(k))
plt.show()
````

````python
from scipy.cluster.hierarchy import linkage, dendrogram

sample_idx = np.random.choice(range(len(X)), size=500, replace=False)
X_sub = X[sample_idx]

link_mat = linkage(X_sub, method='ward')
plt.figure(figsize=(10,6))
dendrogram(link_mat, truncate_mode='level', p=5)
plt.title("Dendrogram for Hierarchical Clustering")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
````

````python
from sklearn.cluster import AgglomerativeClustering
n_clusters = 5
ag = AgglomerativeClustering(n_clusters=n_clusters, linkage='ward')
labels_h = ag.fit_predict(X)

dataset_cleaned_copy['cluster_hier'] = labels_h
````

````python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
````

````python
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names
````

````python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
````

````python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
````

````python
print("Explained variance ratio for 2 components:", pca.explained_variance_ratio_)
print("Total variance explained:", np.sum(pca.explained_variance_ratio_))
````

````python
iris_pca_df = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])
iris_pca_df['target'] = y
````

````python
plt.figure(figsize=(7,5))
sns.scatterplot(data=iris_pca_df, x='PC1', y='PC2', hue='target', palette='Set2', s=70)
plt.title("PCA on Iris Dataset (2 Components)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend(labels=target_names)
plt.grid(True)
plt.show()
````

````python
pca_full = PCA().fit(X_scaled)
plt.figure(figsize=(7,4))
plt.plot(np.cumsum(pca_full.explained_variance_ratio_), marker='o')
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Scree Plot (Iris Dataset)")
plt.grid(True)
plt.show()
````

````python
print("\n===== PART 3: PCA + KMeans on Digits Dataset =====")

from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
import matplotlib.pyplot as plt
import numpy as np
````

````python
digits = load_digits()
X = digits.data
y = digits.target
````

````python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
````

````python
n_components_list = [10, 20, 30, 40, 50, 64]
explained_var = []
ari_scores = []
````

````python
for n in n_components_list:
    pca = PCA(n_components=n, random_state=42)
    X_pca = pca.fit_transform(X_scaled)
    explained_var.append(np.sum(pca.explained_variance_ratio_))
    kmeans = KMeans(n_clusters=10, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_pca)
    ari_scores.append(adjusted_rand_score(y, labels))
````

````python
print(f"n_components={n:2d} | Explained variance={explained_var[-1]:.4f} | ARI={ari_scores[-1]:.4f}")
````

````python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(n_components_list, explained_var, marker='o')
plt.title("Digits: Explained Variance vs PCA Components")
plt.xlabel("Number of PCA Components")
plt.ylabel("Cumulative Explained Variance")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(n_components_list, ari_scores, marker='o', color='orange')
plt.title("Digits: ARI vs PCA Components")
plt.xlabel("Number of PCA Components")
plt.ylabel("Adjusted Rand Index (ARI)")
plt.grid(True)

plt.tight_layout()
plt.show()
````

````python
best_idx = np.argmax(ari_scores)
best_n = n_components_list[best_idx]
best_ari = ari_scores[best_idx]
print(f"\nBest PCA dimension = {best_n}, with ARI = {best_ari:.4f}")
````

````python
pca_best = PCA(n_components=best_n, random_state=42)
X_best = pca_best.fit_transform(X_scaled)
kmeans_best = KMeans(n_clusters=10, random_state=42, n_init=20)
labels_best = kmeans_best.fit_predict(X_best)
````

````python
cluster_centers = kmeans_best.cluster_centers_
rep_indices = []
for k in range(10):
    members = np.where(labels_best == k)[0]
    if len(members) == 0:
        rep_indices.append(None)
        continue
    dists = np.linalg.norm(X_best[members] - cluster_centers[k], axis=1)
    rep_indices.append(members[np.argmin(dists)])

plt.figure(figsize=(10,5))
for i, idx in enumerate(rep_indices):
    plt.subplot(2,5,i+1)
    if idx is None:
        plt.text(0.5,0.5,"Empty",ha='center',va='center')
        plt.axis('off')
    else:
        plt.imshow(digits.images[idx], cmap='gray')
        plt.title(f"Cluster {i}\nTrue:{y[idx]}")
        plt.axis('off')
plt.suptitle(f"Representative Images (PCA={best_n}, ARI={best_ari:.4f})", fontsize=14)
plt.show()
````
